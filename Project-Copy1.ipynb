## Data-Preprocessing

```python
# check the presence of missing value

data.isnull().sum()
```

```python
data
```

```python
# handling the missing values from categorical columns()

mode1=  data['cp'].mode()
mode2=  data['fbs'].mode()
mode3=  data['restecg'].mode()
mode4=  data['exang'].mode()
mode5=  data['thal'].mode()

print('mode of chest_pain_type',mode1)
print()
print('mode of fbs',mode2)
print()
print('mode of restecg',mode3)
print()
print('mode of exang',mode4)
print()
print('mode of thal', mode5)
```

```python
# handling the missing values from categorical columns()

data['cp'].replace(np.nan, ' asymptomatic', inplace=True)
data['fbs'].replace(np.nan, 'False', inplace=True)
data['restecg'].replace(np.nan, 'ST-T wave abnormality', inplace=True)
data['exang'].replace(np.nan, 'no', inplace=True)
data['thal'].replace(np.nan, 'reversable defect', inplace=True)
```

```python
# handling the missing values from numerical column()

data['oldpeak'].fillna(data["oldpeak"].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
# # extraction of indep and dep variable
x= data.iloc[:,0:-1]
y= data.iloc[:,12:13]
```

```python
# independent variables

x
```

```python
# dependent variables

y
```

```python
# handling categorical data

X= pd.get_dummies(x,drop_first=True)
X.head()
```

```python
Y=pd.get_dummies(y,drop_first=True)
Y.head()
```

```python
# y.replace('No disease',0, inplace= True)
# y.replace('Disease',1, inplace= True)
# y
```

```python
# split the dataset

from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test= train_test_split(X,Y,test_size=0.2, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## (I) Logistic Regression Implementation on the Dataset

```python
# model creation

from sklearn.linear_model import LogisticRegression
LRC = LogisticRegression()
LRC.fit(x_train, y_train)
```

```python
# prediction ---test dataset--x_test
y_pred = LRC.predict(x_test)
y_pred
```

### to evaluate the performance of model

```python
# creating confusion matrix

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
# heatmap
import seaborn as sns
sns.heatmap(cm,annot=True,cmap="magma",fmt='.2f')
```

```python
print('TP =', cm[0,0])
print('Tn =', cm[1,1])
print('fp =', cm[1,0])
print('fn =', cm[0,1])
```

```python
# Hard Code for accuracy

acc= (cm[0,0]+cm[1,1])/(cm[0,0]+cm[1,1]+cm[1,0]+cm[0,1])*100
print('accuracy:', acc)
```

```python
# Hard Code for precision

PR = cm[0,0]/(cm[0,0]+cm[1,0])*100
print('precision:',PR)
```

```python
# Hard Code forRecall

RR = cm[0,0]/(cm[0,0]+cm[0,1])*100
print('Recall:', RR)
```

```python
# Hard Code for F1 score

F1score= 2*(PR*RR)/(PR+RR)
print('F1 score', F1score)
```

```python
# inbult method accuracy_score

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```

```python
# complete performance report

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
